### Colocando nossos dados no espaço vetorial






```r
data
```

```r
label <- data[,"Type"]
```

```r
data_cleaned <- data %>% select(-X, -Type)
```

```r
data_train <- data_cleaned[1:150,]
label_train <- label[1:150]
data_test <- data_cleaned[150:nrow(data_cleaned),]
label_test <- label[150:nrow(data_cleaned)]
```

```r
a <- scale(data_train)
attr <- attributes(a)
```

```r

head(data_train)
```

```r
install.packages("e1071")
```

```r
data_train["label"] <- factor(label_train)
```

```r
library("caret")
library("e1071")
df.tree = train(label  ~ ., 
                  data=data_train, 
                  method="rpart2", 
                  trControl = trainControl(method = "cv"))
```

```r
plot(df.tree$finalModel, uniform=TRUE,
     main="Classification Tree")
text(df.tree$finalModel, use.n.=TRUE, all=TRUE, cex=.8)
```

```r
save(attr, file = 'attributes.Rdata')
save(data_test, file="data_test.Rdata")
save(df.tree, file="model.Rdata")
```
